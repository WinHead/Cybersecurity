```python
class Frequency_analysis:
    """
    Программа для осуществления частотного анализа закодированного текста
    """
    
    def __init__(self, file_path):
        '''
        Инициализация класса
        
        Arguments:
        file_path, str -путь к .txt файлу с зашифрованным текстом
        '''
        self.file_path = file_path
        
        # символы кирилицы, расположенные в порядке уменьшения частоты встречаемости 
        self.sorted_russian_letters = [' ', 'о', 'е', 'а', 'и', 'н', 'т', 'с', 'р', 'в', 'л', 'к', 'м', 'д', 'п', 'у', 'я', 'ы', 'з', 'ь', 'б', 'г', 'ч', 'й', 'х', 'ж', 'ю', 'ш', 'ц', 'щ', 'э', 'ф']
        
    def byte_reading(self):
        '''
        Побайтовое считывание файла; Запоминаем частоту появления символов
        '''
        
        with open(self.file_path, 'r') as f:
            # Считываем файл по 1 байту
            symbol = f.read(1)
            
            symbol_counter = 0 # Устанавливаем начальное значение счетчика символов
            symbols_dict = {} # Инициализируем словарь символов
            
            # Пока не считаем пустой символ
            while symbol:
                
                # Инкрементируем счетчик символов
                symbol_counter += 1
        
                # Если символ есть в словаре
                if symbols_dict.get(symbol, 0):
                    # Увеличиваем счетчик появления этого символа
                    symbols_dict[symbol] = symbols_dict.get(symbol) + 1
                else:
                    # Если символа раньше не было, устанавливаем значение счетчика появлений в 1
                    symbols_dict[symbol] = 1
                
                # Считываем очередной сивол
                symbol = f.read(1)
        
            
            # Сортируем символы (ключи словаря) по количеству появлений данного символа в тексте 
            sort_keys = sorted(symbols_dict, key=symbols_dict.get, reverse=True)
            
            symbol_frequency = {} # Инициализируем словарь с частотой символов
            
            # Перебираем все ключи
            for symbol in sort_keys:
                # И считаем частоту появления каждого символа
                symbol_frequency[symbol] = [symbols_dict[symbol], symbols_dict[symbol] / symbol_counter]
                
            self.symbol_frequency = symbol_frequency
            
        return self
            
    def print_frequencies(self):
        '''
        Печать словаря с частотами
        '''
        print(f'<символ> [число появлений, частота]')
        for key, value in self.symbol_frequency.items():
                #print(f'Символ <{key}> встречается в тексте {value[0]} раз с частотой {value[1]}', value)
                print(f'<{key}> {value}')

        
    def decryption(self):
        '''
        Дешифруем зашифрованный текст
        '''
        ''' Сопоставляем символу из текста предполагаемый верный символ '''
        
        decryption_dict = {} # Словарь, где каждому символу из текста ставится в соответствие предполагаемый по частоте символ
        
        counter = 0 # Счетчик символов в словаре
        
        # Извлекаем из отсортированного словаря значения 
        for key, value in self.symbol_frequency.items():
            # Проверяем, не вышли ли за пределы масисва
            if counter > 10:
                break
            # Сопоставляем значения
            decryption_dict[key] = self.sorted_russian_letters[counter]
            
            counter += 1 # инкрементируем счетчик
            
        
        ''' Еще раз считываем файл, параллельно расшифровывая его '''
        print(decryption_dict)
        
        with open(self.file_path, 'r') as f:
            # Считываем файл по 1 байту
            symbol = f.read(1)
            
            # Пока не считаем пустой символ
            while symbol:
                
                print(decryption_dict.get(symbol, symbol), end='')
                
                # Считываем очередной сивол
                symbol = f.read(1)
            
```
